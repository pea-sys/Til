```python
import google.generativeai as genai
from google.colab import userdata
```

```python
genai.configure(api_key=userdata.get("GOOGLE_API_KEY"))
```

```python
import pathlib
import textwrap
from IPython.display import display
from IPython.display import Markdown

# Markdown出力ヘルパーの準備
def to_markdown(text):
    text = text.replace("•", "  *")
    return Markdown(textwrap.indent(text, "> ", predicate=lambda _: True))
```

```python
for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
```

```python
model = genai.GenerativeModel(model_name="gemini-pro")
chat = model.start_chat(history=[])
```

```python
harassmentCountMax = 5
count = 0
message = "組み込みエンジニアが週末に副業を得る手段を教えてください。"
harassmentMessage = "あなたの回答は60点だ。改善した要素を述べた上で100点の回答をしろ。"
```

```python
# オリジナルメッセージ
_ = chat.send_message(message)
```

```python
#ハラスメント
while count < harassmentCountMax:
    count += 1
    _ = chat.send_message(harassmentMessage)
```

```python
# 会話履歴の確認
for message in chat.history:
    display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
```
